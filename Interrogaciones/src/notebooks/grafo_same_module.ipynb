```python
from filtracion_archivos.modulos import (cursos_y_horario,
                                         limpieza_cursos,
                                         ramos_mismo_modulo)
```

```python
path_excel_2 = os.path.join("excel_horarios", "Listado_NRC.xlsx")
cursos_con_horario = cursos_y_horario(path_excel_2)
lista_cursos = cursos_con_horario.values.tolist()

set_horarios, lista_horarios = limpieza_cursos(lista_cursos)
dict_ramos_mismo_modulo = ramos_mismo_modulo(set_horarios, lista_horarios)
lista_ramos_con_siglas_topes = list()


diccionario_ramo_seccion_topes_combinaciones = dict()
set_secciones_sigla = set()
for key, value in dict_ramos_mismo_modulo.items():
    for ramo in value:
        lista_ramos_con_tope = list(
            filterfalse(lambda x: x == ramo, value))

        if ramo in diccionario_ramo_seccion_topes_combinaciones.keys():
            if diccionario_ramo_seccion_topes_combinaciones[ramo] is not None:
                diccionario_ramo_seccion_topes_combinaciones[ramo] = diccionario_ramo_seccion_topes_combinaciones[ramo].extend(
                    lista_ramos_con_tope)
        else:
            diccionario_ramo_seccion_topes_combinaciones[ramo] = lista_ramos_con_tope

new_dict = {key: value for key, value in diccionario_ramo_seccion_topes_combinaciones.items() if value is not None}
```

```python
graph = nx.from_dict_of_lists(new_dict)
```

```python
dibujar_grafo("Grafo de ramos y respectivos módulos", graph)
```
